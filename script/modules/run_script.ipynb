```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
from datetime import datetime
import os
import time
import APTdataloadaer as apt
```

```python
## 경로설정
path = 'C:/Users/sdjhs/git/dataset/DL'

## 날짜 범위 설정
now_month = datetime.now().strftime('%Y-%m')
date_range = pd.date_range(start='2006-01', end=now_month, freq='MS').strftime('%Y%m').tolist()

## 데이터로더 객체 생성
authKey = apt.authKey
dataloader = apt.AptTradingDataloader(authKey)
## 시군구코드 가져오기
sigungu_cd = list(apt.sigungu_cd.values())
```

```python
## date_range.index('201802')
```

```python
for _sigungu in sigungu_cd[2:]:
    print(f"시군구코드: {_sigungu} 데이터 수집 시작")
    for _date in date_range[:]:
        raw = dataloader.get_apt_trading_raw(_sigungu, _date)
        dataset, df = dataloader.preproc_apt_trading(raw)

        dir_path = os.path.join(f"{path}/trading", _sigungu)
        if not os.path.exists(dir_path):
            os.makedirs(dir_path)

        df.to_csv(os.path.join(dir_path, f'{_date}.csv'), index=False, encoding='utf-8-sig')

        detail_path = os.path.join(f"{path}/trading_detail", _sigungu)
        if not os.path.exists(detail_path):
            os.makedirs(detail_path)

        dataset.to_csv(os.path.join(detail_path, f'{_date}.csv'), index=False, encoding='utf-8-sig')

        time.sleep(3)
```
